PSC dataloader's batch 20 & data's batch 50


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import torch.nn as nn
import sklearn
from sklearn import metrics
from sklearn.cluster import KMeans
from psc import PSC, Accuracy
import time
import torch
import random
import warnings
from scipy.io import arff
import os

warnings.filterwarnings("ignore")

torch.manual_seed(0)
np.random.seed(0)
random.seed(0)
os.environ["PYTHONHASHSEED"] = str(0)

data = arff.loadarff('connect-4.arff')
df = pd.DataFrame(data[0])
# df['class'] = df['class'].astype(int)
print(df.info())
print(df.head(5))

y = df['class'].values
print(y)
print("y shape ", y.shape)
x_data = df.drop(['class'], axis = 1).values

scaler = sklearn.preprocessing.StandardScaler().fit(x_data)
x = scaler.transform(x_data)
print(x)
print("x shape ", x.shape)

class Net1(nn.Module):
    def __init__(self, out):
        super(Net1, self).__init__()
        self.output = out
        # Define the layers
        self.fc1 = nn.Linear(42, 64)
        self.fc2 = nn.Linear(64, 128)
        self.fc3 = nn.Linear(128, 64)
        self.output_layer = nn.Linear(64, self.output)

        self.relu = nn.ReLU()

    def forward(self, x):

        x = self.fc1(x)
        x = self.fc2(x)
        x = self.fc3(x)
        x = self.output_layer(x)
        return x
```

```python
from sklearn.utils import check_random_state

print(check_random_state)
```

```python
i = 3
print(f"n_clusters = {i}")
model = Net1(out=i)
kmeans = KMeans(n_clusters=i, init='random', n_init='auto', algorithm='elkan')
psc = PSC(model=model, clustering_method=kmeans, test_splitting_rate=0, n_neighbor=i, epochs=50, batch_size_data=50, batch_size_dataloader=20)
time1 = round(time.time()*1000)
x_fit_psc = psc.fit(x)
x_predict_psc = psc.predict(x)
time2 = round(time.time()*1000)
print(f"PSC time spent: {time2 - time1} milliseconds")

# print(f"silhouette: {metrics.silhouette_score(x, x_fit_psc.labels_)}")
# print(f"calinski_harabasz: {metrics.calinski_harabasz_score(x, x_fit_psc.labels_)}")
# print(f"homogeneity: {metrics.homogeneity_score(y, x_predict_psc)}")
# print(f"v_measure: {metrics.v_measure_score(y, x_predict_psc)}\n")s
acc = Accuracy(y_true=y, y_pred=x_predict_psc)
acc.acc_report()
```
